# Example 4.14: Whale Sightings

Ecologists often use the number of reported sightings of a rare species of animal to estimate the remaining population size. For example, suppose the number, x, of reported sightings per week of blue whales is recorded. Assume that x has approximately a Poisson distribution with a mean of 2.6 per week. 
1. Find the mean and standard deviation of x. Interpret the result.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as sts
%matplotlib qt
plt.close(fig='all')

lamb = 2.6
mu = lamb
sig = np.sqrt(lamb)

print('E(x) = ', mu)
print('sigma = ', sig)

# note the ceil and floor as we can only have discrete values
print('lowerborder (value included)', np.ceil(mu-2*sig))
print('upperborder (value included)',np.floor(mu+2*sig))
```

2.Let's look at the pmf and cdf

```python
# Frozen Poisson Distribution with lambda = 2.6

rv = sts.poisson(lamb)
# Let's look at the first 11 discrete values
x = np.arange(0,11)
plt.bar(x,rv.pmf(x))
plt.plot(x,rv.cdf(x),'o--')
plt.title('Poisson distribution with $\lambda$ = 2.6')
plt.xlabel('random variables')
plt.ylabel('probability or cummulative ')
plt.legend(('cdf','pmf'))
plt.show()
```

 3. P(x<2) ?

```python
print('P(x < 2) =',rv.cdf(1))
```

4. P(x>5)

```python
print('P(x > 5) =',1-rv.cdf(5))
```

5. P(x = 5)?

```python
print('P(x = 5) =',rv.pmf(5))
```
