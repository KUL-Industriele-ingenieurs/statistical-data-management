Suppose a poll of 20 voters is taken in a large city. The purpose is to determine the numbers who favor a certain candidate for mayor. Suppose that 60% of all the city‚Äôs voters favor the candidate. Find the following: 



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as sts
%matplotlib inline
```


i.Find the mean and standard deviation of x.  ùúá=20√ó0.6=12  and  ùúé2=20√ó0.6√ó0.4=4.8  and  ùúé=2.19

```python
p = 0.6
n = 20
mu = n*p
sig = np.sqrt(n*p*(1-p))

print('mu = ', mu)
print('sigma = ', sig)
```

Calculate the Probability Mass Function and the Cummulative Distribution.
Also show both.


```python
x = np.arange(0,n+1)
print(x)
# Frozen random variable distribution
rv = sts.binom(n,p)
print(rv)
plt.bar(x,rv.pmf(x))
plt.plot(x,rv.cdf(x),'o--')
plt.title('Binomial distribution with n = 20 and p = 0.6')
plt.xlabel('random variables')
plt.ylabel('probability or cummulative ')
plt.legend(('cdf','pmf'))
plt.show()
```

ii.Find the probability that x‚â§10.

```python
print('P(x=<10) =',rv.cdf(10))
```

iii.Find the probability that x>12. 

```python
print('P(x>12) =',1-rv.cdf(12))
```

iv.Find the probability that x=11

```python
print('P(x = 11) =',rv.pmf(11))
```

v.Interval $\mu \pm 2\sigma$

```python
# note the ceil and floor as we can only have discrete values
print('lowerborder (value included)', np.ceil(mu-2*sig))
print('upperborder (value included)',np.floor(mu+2*sig))
print(mu-2*sig)
print(mu+2*sig)
```
