The heart association claims that only 10% of US adults over 30 years of age meet the minimum fitness requirements established by the President's Council on fitness. Suppose 4 adults are randomly selected and each is given the fitness test.

a) use the formula for the binomial random variable to find the probability distribution of x, the number of adults who pass the fitness test.

Initialisation: 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as sts
%matplotlib inline
```

Create the probability Mass Function.
Add print to view the content of the variables.
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.binom.html
https://docs.scipy.org/doc/numpy/reference/generated/numpy.arange.html


```python
# ranging form 0->4
x = np.arange(0,5)
print(x)
# Frozen random variable distribution
rv = sts.binom(4,0.1)
```

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.binom.html


```python
plt.bar(x,rv.pmf(x))
print(rv.pmf(x))
#plt.hold(True)
plt.plot(x,rv.cdf(x),'o--')
plt.title('Binomial distribution with n = 4 and p = 0.1')
plt.xlabel('random variables')
plt.ylabel('probability ')
plt.legend(('cdf','pmf'))
plt.show()
```
