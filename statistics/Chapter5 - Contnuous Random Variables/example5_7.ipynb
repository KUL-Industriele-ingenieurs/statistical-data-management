# Example 5.7: Cell Phone Application

Assume that the length of time x, between charges of a cellular phone is normally distributed with a mean of 10 hours and a standard deviation of 1.5 hours.

Find the probability that the cell phone will last between 8 and 12 hours between charges.

See:
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html#scipy.stats.norm
https://docs.scipy.org/doc/numpy/reference/generated/numpy.linspace.html



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as sts
%matplotlib inline


# Frozen random variable distribution
rv =  sts.norm(loc = 10,scale = 1.5)
x = np.linspace(rv.ppf(0.01),rv.ppf(0.99), 100)
print(x)
```

Show the pdf and cdf.

```python
plt.plot(x,rv.pdf(x))
plt.title('pdf of normal distribtution with $\mu$ = 10 and $\sigma$ = 1.5')
plt.xlabel('random variables x')
plt.ylabel('probability density function pdf(x)')

plt.show()

plt.figure()
plt.plot(x,rv.cdf(x),'r--')
plt.title('cdf of normal distribtution with $\mu$ = 10 and $\sigma$ = 1.5')
plt.xlabel('random variables x')
plt.ylabel('cumulative density function cdf(x)')

plt.show()
```

P(8<x<12)?

```python
x1 = 8
x2 = 12
print('P(x < 12) - P(x < 8) =',rv.cdf(x2)-rv.cdf(x1))
```

Solving the example with the standard normal distribution

```python
# standard normal distribution is the default setting
rv2 = sts.norm()

z1 = (x1 - 10)/1.5
z2 = (x2 - 10)/1.5


print('z-score for x= ',x1,' is ',z1)
print('z-score for x= ',x2,' is ',z2)

print('P(z < -1.333) =', rv2.cdf(z1))
print('P(-1.333< z < 1.333) =', 1 - 2*rv2.cdf(z1))
```
