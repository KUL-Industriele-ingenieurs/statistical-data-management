# Example 5.11: College Entrance Exam Application

Suppose the scores, x, on a college entrance examination are normally distributed with a mean of 550 and a standard deviation of 100. A certain prestigious university will consider for admission only those applicants whose scores exceed the 90nd percentile of the distribution. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as sts
%matplotlib inline



# Frozen random variable distribution
rv =  sts.norm(loc = 550,scale = 100)
x = np.linspace(rv.ppf(0.01),rv.ppf(0.99), 100)
print(rv.ppf(0.01))
print(rv.ppf(0.99))


plt.plot(x,rv.pdf(x))
plt.title('pdf of normal distribtution with $\mu$ = 550 and $\sigma$ = 100')
plt.xlabel('random variables x')
plt.ylabel('probability density function pdf(x)')

plt.show()

plt.figure()
plt.plot(x,rv.cdf(x),'r--')
plt.title('cdf of normal distribtution with $\mu$ = 550 and $\sigma$ = 100')
plt.xlabel('random variables x')
plt.ylabel('cumulative density function cdf(x)')

plt.show()
```

Find the minimum score an applicant must achieve in order to receive consideration for admission to the university.

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.rv_continuous.ppf.html



```python
print('P(x < ?)  = 0.9 => ',rv.ppf(0.9))
```

Using the standard normal distribution.

```python
# Frozen standard normal distribution
rv2 = sts.norm()
# threshold value zt for standard normal distribution
zt = rv2.ppf(0.9)
print('P(z < ?)  = 0.9 => ',zt)
xt = rv.mean() + rv.std() * zt
print('P(x < ?)  = 0.9 => ',xt)
```
