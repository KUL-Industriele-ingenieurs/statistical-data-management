# Example 7.2: Mean Hospital Length of Stay, $\sigma$ unknown with Pandas
Consider a large hospital that wants to estimate the average length of stay of its patients. The hospital randomly samples 100 of its patients and find that the sample average length of stay is 4.53 days.  What is the 95% confidence interval for the average length of stay of all the patients? 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as sts
#%matplotlib qt
#plt.close('all')
%matplotlib inline


df = pd.read_csv('HOSPLOS.CSV')
print(df)
plt.title('Length of Stay')
plt.boxplot(df['LOS'])
plt.ylabel('Days')
plt.show()
```

Find the sample mean, sample standard deviation and standard error of the sample mean.

```python
xb = df['LOS'].mean()
s = df['LOS'].std(ddof = 1)
n = len(df)
sm = s / np.sqrt(n)
print('sample mean: ',xb)
print('standard devitation sample (unbiased): ',s)
print('number of observations in the sample: ',n)
print('standard devitation sample mean (standard error of the mean): ',sm)
```

Find z1 such that  P(-z1<z<z1) = 95%

Freeze the parameters of the Standard normal distribution


```python
rv = sts.norm()
z1 = rv.ppf(0.95+0.025)
print('z1: ',z1)
```

95% confidence interval for the mean



```python
print('95% confidence interval ', [xb - z1 * sm , xb + z1 * sm])
print('sampling error: ', z1 * sm)
```

Sample size for a sampling error of 0.25 days (sigma unknown)

```python
SE = 0.25
n_new = np.ceil((z1*s/SE)**2)
print('new sample size: ',n_new)
```
