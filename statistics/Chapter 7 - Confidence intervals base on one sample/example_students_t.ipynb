# Example student t test
We will use different degrees of freedom and illustrate the pdf and the cdf.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as sts
#%matplotlib qt
#plt.close('all')
%matplotlib inline


# make a vector of t values
t = np.linspace(-7,7, 100)

# initialize two figures
#fig1, ax1 = plt.subplots()
#fig2, ax2 = plt.subplots()
```

Loop over different degrees of freedom.

```python
fig1, ax1 = plt.subplots()
for i in [1, 2, 3, 5, 10, 100]:
    p = sts.t.pdf(t,i)
    ax1.plot(t,p, label = str(i))
    ax1.legend()
    

plt.title('pdf of students t distribution for different degree of freedom')
plt.xlabel('random variables t')
plt.ylabel('probability density function pdf(t)')
plt.show()

fig2, ax2 = plt.subplots()
for i in [1, 2, 3, 5, 10, 100]:
    c = sts.t.cdf(t,i)
    ax2.plot(t,c, label = str(i))
    ax2.legend()


plt.title('cdf of students t distribution for different degree of freedom')
plt.xlabel('random variables t')
plt.ylabel('probability density function cdf(t)')
plt.show()
```
