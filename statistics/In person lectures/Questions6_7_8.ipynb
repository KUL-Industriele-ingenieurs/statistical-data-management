```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as sts
#%matplotlib qt
#plt.close(fig='all')

%matplotlib inline
```

Multiple choice question: Consider a test with $H_0: \mu = 4$ and $H_a: \mu \lt 4$ consider $\sigma = 1$. Give the rejection region using the z-statistic and $\alpha = 0.08$.

```python
# Freeze the parameters of the standard normal distribution
rv = sts.norm()
# One-tailed test with rejection region
z1 = rv.ppf(0.08)
print('rejection region < ',z1)


# Plot the standard normal distribution

fig1 = plt.figure()
zz = np.linspace(rv.ppf(0.0001),rv.ppf(0.9999), 100)
plt.plot(zz,rv.pdf(zz))
plt.title(r'Rejection region')
plt.xlabel(r'random variables $z$')
plt.ylabel('pdf of standard normal distribution')

# color the reject region
part = np.linspace(rv.ppf(0.0001),z1, 100)
plt.fill_between(part,rv.pdf(part))
plt.show()
```

Multiple-choice question: If the proportion of pupils having a tablet differs from 25%, then changes in policy are required. Suppose a hypothesis test is conducted and the test statistic is z=2.4. Find the p-value.

```python
# Freeze the parameters of the standard normal distribution
rv = sts.norm()
# two-tailed observed significance level
p = 2*rv.cdf(-2.4)
print('The observed significance level', p)
fig1 = plt.figure()
zz = np.linspace(rv.ppf(0.0001),rv.ppf(0.9999), 100)
plt.plot(zz,rv.pdf(zz))
plt.title(r'Observed significance level')
plt.xlabel(r'random variables $z$')
plt.ylabel('pdf of standard normal distribution')

# color the surface area is observed significance level
part1 = np.linspace(2.4,rv.ppf(0.9999), 100)
part2 = np.linspace(rv.ppf(0.0001),-2.4, 100)
plt.fill_between(part1,rv.pdf(part1))
plt.fill_between(part2,rv.pdf(part2))

plt.show()
```

Open Question: Due to certain limitations in the graphical user interface (GUI) 3.0, the engineers of Netflix made an updated version called GUI 4.0. A sample of 171 randomly selected Netflix clients exposed to both GUIs was asked which method they preferred. As a result, 138 clients indicated their preference for GUI 4.0. To roll out GUI 4.0, the management board at Netflix requires that more than 70 % of the clients prefer the new GUI. Should GUI 4.0 be rolled out?  Also, mention all assumptions that need to be fulfilled to make this test. Assume a significance level of 0.05.

$H_0 : p = p_0 = 0.7$ 

$H_a : p > p_0 $

    

```python
# set some initial values
n = 171
success = 138
alpha = 0.05
p_hat = success/n
p_0 = 0.7
q_0 = 0.3
```

Conditions required for a valid large-sample hypothesis test for p:
1. A random sample from a binomial population => okay
2. Large sample because $p_0 n  \geq 15$ and $q_0 n \geq 15$ => okay

```python
# One-tailed test construction of rejection region
# Freeze the parameters of the standard normal distribution
rv = sts.norm()
# One-tailed test with rejection region
z1 = rv.ppf(1-alpha)
print('rejection region > ',z1)


# Plot the standard normal distribution

fig1 = plt.figure()
zz = np.linspace(rv.ppf(0.0001),rv.ppf(0.9999), 100)
plt.plot(zz,rv.pdf(zz))
plt.title(r'Rejection region')
plt.xlabel(r'random variables $z$')
plt.ylabel('pdf of standard normal distribution')

# color the reject region
part = np.linspace(z1,rv.ppf(0.9999), 100)
plt.fill_between(part,rv.pdf(part))
plt.show()
```

```python
# calculate the large sample statistic
z = (p_hat - p_0)/np.sqrt(p_0*q_0/n)
print ('large sample statistic',z)
print ('statistic in rejection region ? ', z>z1)
```

Conclusion: We reject $H_0$ in favor of $H_a$ at a level of significance of $alpha = 0.05$.
Hence GUI 4.0 can be rolled-out.
