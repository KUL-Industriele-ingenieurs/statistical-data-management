# Multiple Choice Question: 
Estimate the difference between two population means correctly within Â±0.5 with a confidence level of 95%. Assume a equal variances of 20 for both populations. What is the sample size for each of the population?

```python
# Let's assume first large sample statistics n1 and n2 is 30 or larger
# Calculate z_{\alpha/2} for 95% confidence level
alpha = 1-0.95
# Freeze the parameters of the standard normal distribution
rv = sts.norm()
# Calculate z_{\alpha/2}
z1 = rv.ppf(1-alpha/2)
print('z_{alpha/2} : ',z1)

# Assume same size samples from both populations n_1 = n_2 = n
# The population variances are equal var1 = var2 = var = 20
# Sampling error is 0.5
SE = 0.5
var = 20
n = np.ceil((z1**2 * (2*var))/SE**2)
print ('n_1 = n_2 = ', n)    

```

# Open Question: 
A pharmaceutical company wants to determine if the company's new medicine for flu is effective after two days in babies. Among the 120 babies who had the flu and were given the medicine, 29 were cured within 2 days. Among the 280 babies who had the flu but were not given the medicine, 56 recovered within two days. Use alpha = 0.01.

The pharmaceutical company claims that the medicine tackles the flu in babies in the first two days. Perform a test of hypotheses to inspect the company's claim.  

$p_1$ : population fraction cured within 2 days with medicine (treatment) 

$p_2$ : population fraction recovcered within 2 days without medicine (control)

$H_0 : (p_1 - p_2) = 0 $ 

$H_a : (p_1 - p_2) > 0 $

It is a one-tailed test. 




```python
# Babies given medication
n1 = 120
# Babies cured
x1 = 29
p_hat1 = x1 / n1
q_hat1 = 1- p_hat1
# Babies without medication 
n2 = 280
#Babies recovered
x2=56
p_hat2 = x2 / n2
q_hat2 = 1- p_hat2
```

Check if large sample conditions apply.
If we assume that the babies are randomly selected in an indepedent manner then we still have to check if the samples sizes fullfil the large sample requirements:

```python
# Check assumptions large sample
print ('n1 * p_hat1 >= 15 :', n1 * p_hat1 >= 15)
print ('n1 * q_hat1 >= 15 :', n1 * q_hat1  >= 15)
print ('n2 * p_hat2 >= 15 :', n2 * p_hat2 >= 15)
print ('n2 * q_hat2 >= 15 :', n2 * q_hat2  >= 15)
```

We use the large-sample test and hence the z-statistic. Let's find $ z_{\alpha}$ such that $P(z > z_{\alpha}) = \alpha = 0.01$.

```python
# Freeze the parameters of the standard normal distribution
rv = sts.norm()
# Calculate z_{\alpha}
alpha = 0.01
z1 = rv.ppf(1-alpha)
print('z_{alpha} : ',z1)
```

```python
# Calculate the test-statistic
# Assume that H_0 is true we then can make a better estimate of p_hat based on both samples
p_hat = (x1 + x2) / (n1 + n2)
q_hat = 1 - p_hat
z = (p_hat1 - p_hat2) / np.sqrt(p_hat*q_hat * (1/n1 + 1/n2)) 
print ('The test statistic is : ', z)
print('In rejection region? ',z > z1)
```

Conclusion:
There is insufficient evidence to reject $H_0$, both population proporations are equal, in favor for $H_a$ where the population proportion cured with the medicine is larger than the population proportion recovered without medicine and  with a significance level of 0.01. 
    
    
    
    

# Open Question: 
The diameter of steel rods manufactured on two different extrusion machines is being investigated. Two random samples of size ð‘›_1=15 and ð‘›_2=17 are selected, and the sample means in mm and sample variances are $\bar{x}_1 =8.73$, $ð‘ _1^2=0.35$ , and $\bar{x}_2=8.68$, $ð‘ _2^2=0.40$, respectively. Assume that the data are drawn from normal distributions. Use ð›¼=0,05. 

Set up a test to compare the variance of both samples. What is your conclusion?


$H_0 : \sigma_1^2 / \sigma_2^2 = 1 $ 

$H_a : \sigma_1^2 / \sigma_2^2 \neq 1 $

It is a two-tailed test. 
Conditions required:
1. The samples are random and independent.
2. Both populations are normally distributed

Set up a rejection region.  We force a one-tailed test by putting the largest sample variance in de numerator and the smallest sample variance in the denominator.

Hence F(numerator df 17-1, denominator df 15-1)


```python
# Set up rejection region

n1 = 15
n2 = 17
alpha = 0.05
rv1 = sts.f(n2-1,n1-1)
ff =  rv1.ppf(1-alpha/2)
print ('F_alpha/2 ',ff) 
```

```python
# Test statistic
v1= 0.35
v2= 0.40
F = v2/v1
print('F-statistic: ',F)
print('in rejection region? ',F > ff)
```

Conclusion: Insufficient evidence to reject $H_0$, population variances are equal, in favor of $H_a$ , population variances are different with alpha = 0.05.

Construct a confidence interval for the difference in mean rod diameter. Give a practical conclusion of your findings useful for the manufacturer.

Conditions required for a valid small-sample confidence interval for  $\mu_1 - \mu_2$:
1. Two samples are randomly selected in an independent manner from the two target populations
2. Both sampled populations have distributions that are approximately normal.
3. The population variances are equal. 



```python
# Constructing the confidence interval
alpha = 0.05
rv = sts.t(n1+n2-2)
tt =  rv.ppf(1-alpha/2)
print ('t_alpha/2 ',tt) 
# Pooled-variance as estimate of the variance of the population
varp =  ((n1-1)*v1 + (n2-1)*v2)/(n1+n2-2)
print ('Pooled-variance: ', varp)
# estimate of the standard error
st_er = np.sqrt(varp*(1/n1 + 1/n2))
print ('std sampling distribution (standard error) ',st_er)

# sampling error
SE = tt * st_er
print('Sampling error: ',SE)
#%% 95% Confidence interval 
xb1 = 8.73
xb2 = 8.68
print('95% confidence interval: ',[(xb1 - xb2 - SE), (xb1 - xb2 + SE)])
```

There is insufficient evidence to indicate that $\mu_1 - \mu_2$ differs from zero because the 95% confidence interval includes zero as possible value for $\mu_1 - \mu_2$.
