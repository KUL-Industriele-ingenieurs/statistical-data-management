## Example 9.1
A dietician developed a diet that is low in fats, carbohydrates and cholesterol (Diet A). Two random samples of 100 obese people each were selected and one group of 100 was placed on Diet A. The other 100 were placed on a diet that contained approximately the same quantity of food, but was not as low in fats, carbohydrates and cholesterol (Diet B). The amount of weight loss in a 3-week period was recorded. 
1. Form a 95% confidence interval for the difference between the population mean weight losses for the two diets. 



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as sts
%matplotlib inline
plt.close('all')
df = pd.read_csv('DIETSTUDY.CSV')
print(df)

# We make a new dataframe with selecting the row that are NEW in the column METHODS and the column SCORE.
df1 = df.loc [df['DIET'] =='LOWFAT', ['WTLOSS']]
df2 = df.loc [df['DIET'] =='REGULAR', ['WTLOSS']]
print(df1)
print(df2)
```

Calculate the mean and the standard deviation of x1 samples.

```python
xb1 = df1['WTLOSS'].mean()
s1 = df1['WTLOSS'].std(ddof = 1)
n1 = len(df1)
sm1 = s1 / np.sqrt(n1)
print('sample mean: ',xb1)
print('standard devitation sample (unbiased): ',s1)
print('number of observations in the sample: ',n1)
print('standard devitation sample mean (standard error of the mean): ',sm1)
```

Calculate the mean and the standard deviation of x2 samples.

```python
xb2 = df2['WTLOSS'].mean()
s2 = df2['WTLOSS'].std(ddof = 1)
n2 = len(df2)
sm2 = s2 / np.sqrt(n2)
print('sample mean: ',xb2)
print('standard devitation sample (unbiased): ',s2)
print('number of observations in the sample: ',n2)
print('standard devitation sample mean (standard error of the mean): ',sm2)
```

Calculate the boundaries of the rejection region.

```python
# freeze the random variable to standard normal
rv = sts.norm()
zz = rv.ppf(0.95+0.025)
print('zz: ',zz)
```

Standard Error of $\bar{x}_1 - \bar{x}_2$

```python
sig = np.sqrt(s1**2/n1 + s2**2/n2)
print ('Standare Error   = ',sig)
```

Sampling Error and 95% confidence interval in Pounds ( = 0.45 kg)

```python
SE = zz * sig
print ('Sampling Error = ', SE)
print('95% confidence interval = ',[(xb1 - xb2 - SE), (xb1 - xb2 + SE)])
```

Note that 0 is not included in the confidence interval.
Hence we can conclude that  the mean weight loss induced by Lower fat diet is larger than the weight loss induced by regular diet. This conclusion is obtained with a 95% confidence interval.

$H_0 : (\mu_1 - \mu_2) = 0 $ 

$H_a : (\mu_1 - \mu_2) \neq 0 $

It is a two-tailed test. 

2. What is the rejection region?

As $n \geq 30$ we use the large-sample test and hence the z-statistic. Let's find $ z_{\alpha/2}$ such that $P(|z| > z_{\alpha/2}) = \alpha = 0.05$.


```python
print('zz: ',zz)
```

$P(|z| > 1.96) = \alpha = 0.05$

4. Calculate the z-statistic.

```python
D0 = 0
z = (xb1 - xb2 - D0) / sig
print('z-statistic: ',z)
print('in rejection region? ',np.abs(z) > zz)
```

In conclusion we can state  that we reject the null hypothesis in favor of the alternative hypothesis at an alpha 0,05  level of significance.
