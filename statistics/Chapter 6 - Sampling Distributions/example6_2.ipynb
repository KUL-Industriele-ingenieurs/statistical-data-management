# Example 6.2
The rolling machine of a steel manufacturer produces sheets of steel of varying thickness. The thickness of a steel sheet follows a uniform distribution with values between 150 mm and 200 mm. Supose we perform the following experiment over and over again: randomly sample 11 steel sheets from the production line and record the thickness x of each. Calculate th two sample statistics: sample mean and median.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as sts
#%matplotlib qt
#plt.close(fig='all')
%matplotlib inline
```

Simulate data drawn for a uniform distribution:
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.uniform.html#scipy.stats.uniform
and
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.rv_continuous.rvs.html



```python
# Random Variable Distribution [loc loc+scale]
rv = sts.uniform(loc = 150, scale = 50)
# X is a 1000 x 11 matrix 
# each row is one sample
X = rv.rvs([1000,11])
print(X[0:5,:])
```

Plot historgram of the first sample.

```python
#%% plot histogram for the first column
plt.hist(X[:,0],10)
plt.title('uniform distribution')
plt.xlabel('thickness')
plt.ylabel('absolute frequency')
plt.show()
```

Calculate the mean of each sample and plot the histogram of the 1000 sample means.

```python
fig, ax = plt.subplots()
# xb (xbar) is a 1000 x 1 vector
xb = np.mean(X,axis = 1)
print(xb.shape)
plt.hist(xb,10)
plt.title('distribution of sample mean of uniform distribution')
plt.xlabel('sample mean of thickness')
plt.ylabel('absolute frequency')
plt.show()
```

Calculate the median of each sample and plot the histogram of the 1000 sample means.

```python
fig, ax = plt.subplots()
m = np.median(X,axis = 1)
plt.hist(m,10)
plt.title('distribution of sample median of uniform distribution')
plt.xlabel('sample mean of thickness')
plt.ylabel('absolute frequency')
plt.show()
```

Print some inportant values of the distribution.

```python
# calcualte the mean/std of all data points. Hence 11*1000 = 11000
print('mean of the uniform distribution (approx with n  = 11000) ', np.mean(X))
print('std of the uniform distribution (approx with n  = 11000)', np.std(X,ddof = 1))

# calcualte the mean/std of all the sample means. Hence 1000.
print('mean of the distribution of the sample mean (approx with n  = 1000) ', np.mean(xb))
print('std of distribution of the sample mean (approx with n  = 1000)', np.std(xb, ddof = 1))
# calcualte the mean/std of all the sample median. Hence 1000.
print('mean of the  distribution of the sample median (approx with n  = 1000) ', np.mean(m))
print('std of distribution of the sample median (approx with n  = 1000) ', np.std(m, ddof = 1))
```
