# Example 6.7: Using the Central Limit Theorem to Find a Probability
Suppose we have selected a random sample n = 36 observations from a population with mean equal to 80 and standard diviation equal to 6. The population is not normally distributed.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as sts
#%matplotlib qt
#plt.close(fig='all')
%matplotlib inline
```

We freeze the distribution to a standard error of the means of a sample with $\mu = \mu_{\bar{x}} =  80$ and $\sigma_{\bar{x}} = \sigma / \sqrt{x} = 6 / \sqrt{36}$.

```python
# Freeze the parameters of the normal distribution
rv = sts.norm(loc = 80, scale = 1)
xb = np.linspace(rv.ppf(0.001),rv.ppf(0.999), 100)
```

Let's plot the pdf.
Note the Latex notation below.
https://matplotlib.org/3.1.3/tutorials/text/mathtext.html

```python
plt.plot(xb,rv.pdf(xb))
plt.title(r'pdf of sampling distribtution with $E(\bar{x})$ = 80 and $\sigma_{\bar{x}}$ = 1')
plt.xlabel(r'random variables $\bar{x}$')
plt.ylabel('probability density function (pdf)')
part = np.linspace(82,rv.ppf(0.999), 100)
plt.fill_between(part,rv.pdf(part))
plt.show()
```

Now let's plot the cdf.

```python
plt.figure()
plt.plot(xb,rv.cdf(xb),'r--')
plt.title(r'cdf of sampling distribtution with $E(\bar{x})$ = 80 and $\sigma_{\bar{x}}$ = 1')
plt.xlabel(r'random variables $\bar{x}$')
plt.ylabel('cumulative density function (cdf)')

plt.show()




```

Now calcualte $P(\bar{x} > 82)$

```python
print(r'P($\bar{x}$ > 82) = ', 1 - rv.cdf(82))
```
